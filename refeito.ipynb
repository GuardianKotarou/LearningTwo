````python
#0) Seu código aqui

url1 = "https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl"
j = requests.get(url1)
j.status_code
data_json1 = j.json()

base_dados_1 = pd.DataFrame(data_json1['items'])
base_dados_1
````

````python
#1) Seu código aqui
municipio = pd.DataFrame(base_dados_1['interessado'])
estado = pd.DataFrame(base_dados_1['uf'])

municipio = municipio.rename(columns={'interessado': 'Municípios'})
estado = estado.rename(columns={'uf': 'Estados'})
tabela = pd.concat([estado, municipio], axis=1)

tabela
````

````python
# 2) Seu código aqui

MG = base_dados_1.loc[base_dados_1['uf'] == 'MG']
mgdata = pd.DataFrame(MG)

mgdatab = mgdata[mgdata['tipo_interessado'] == 'Município']
mgdatan = pd.DataFrame(mgdatab)
mgdatai = mgdatan[mgdatan['status'] == 'Arquivado por decurso de prazo']['uf'].value_counts()
tabela = pd.DataFrame(mgdatai)
tabela = tabela.rename(columns={'interessado': 'Deferidos'})


tabela
````
